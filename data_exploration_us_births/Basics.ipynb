```python
def read_csv(filename):
    f = open(filename, 'r')
    readfile = f.read()
    list_newline = readfile.split('\n')
    string_list = list_newline[1:]   
    final_list = []
    for each in string_list:
        int_fields = []
        string_fields = each.split(',')
        for x in string_fields:
            a = int(x)
            int_fields.append(a)
        final_list.append(int_fields)
    return final_list
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
print(cdc_list[:10])

def month_births(list_list):
    births_per_month = {}
    for lists in list_list:
        births_per_month[lists[1]] = births_per_month.get(lists[1], 0) + lists[4]
    return births_per_month
filename = "US_births_1994-2003_CDC_NCHS.csv"
#cdc_month_births = month_births(read_csv(filename))  
#print(cdc_month_births)

def dow_births(list_list):
    births_per_dayweek = {}
    for lists in list_list:
        births_per_dayweek[lists[3]] = births_per_dayweek.get(lists[3], 0) + lists[4]
    return births_per_dayweek
filename = "US_births_1994-2003_CDC_NCHS.csv"
cdc_day_births = dow_births(read_csv(filename))  
print(cdc_day_births) 

def calc_counts(data, column):
    births_per_column = {}
    for lists in data:
        births_per_column[lists[column]] = births_per_column.get(lists[column], 0) + lists[4]
    return births_per_column
filename = "US_births_1994-2003_CDC_NCHS.csv"
cdc_day_births = calc_counts(read_csv(filename), 1)  
print(cdc_day_births) 
    
```

```python
def list_splitcomma(file):
    f = open(file, 'r')
    f_splitline = f.read().split('\n')[1:]
    f_final = []
    for each in f_splitline: 
        f_splitcomma = each.split(',')
        f_final.append(f_splitcomma)  
    return f_final

def appendfile(file, content):
    f = open(file, 'a')
    f.write('\n')
    f.write(content)

def copyfile(file1, file2):
    dict2 = list_splitcomma(file2)
    for each in dict2:
        if each[0] in ['2000', '2001', '2002', '2003']:
            a = each[0] + 'b' + ',' + each[1] + ',' + each[2] + ',' + each[3] + ',' + each[4]  
            #print(a)
            appendfile(file1, a)
        else:
            a = each[0] + ',' + each[1] + ',' + each[2] + ',' + each[3] + ',' + each[4]  
            #print(a)
            appendfile(file1, a)       
    return file1
file1 = 'US_births_1994-2003_CDC_NCHS.csv'
file2 = 'US_births_2000-2014_SSA.csv'
copyfile(file1, file2)
a = open('US_births_1994-2003_CDC_NCHS.csv','r' )
print(a.read())






```

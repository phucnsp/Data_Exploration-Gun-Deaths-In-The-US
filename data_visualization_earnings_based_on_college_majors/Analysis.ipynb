# Pandas, Bar Plots

```python
import numpy as np
recent_grads['Median'] = recent_grads['Median'].sort_values(ascending=False, na_position='last')
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.bar(np.arange(10)+0.5,recent_grads[:10]['ShareWomen'])
ax1.set_xlim(0,11)
ax1.set_title(' Women rate in 10 highest paying majors ')

ax2.bar(np.arange(10)+0.5,recent_grads[-11:-1]['ShareWomen'])
ax2.set_xlim(0,11)
ax2.set_title(' Women rate in 10 lowest paying majors ')
```

```python
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.bar(np.arange(10)+0.5,recent_grads[:10]['Unemployment_rate'])
ax1.set_xlim(0,11)
ax1.set_title(' Unemployment_rate in 10 highest paying majors ')

ax2.bar(np.arange(10)+0.5,recent_grads[-11:-1]['Unemployment_rate'])
ax2.set_xlim(0,11)
ax2.set_title(' Unemployment_rate in 10 lowest paying majors ')
```

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[-11:-1].plot.bar(x='Major', y='ShareWomen', legend=False)
```
